## üèµÔ∏èEDA dan Visualisasi

```python
# Visualisasi 1: Distribusi Usia (Age)
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=20, kde=True)
plt.title('Distribusi Usia Pelanggan')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=data)
plt.title('Jumlah Pelanggan Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

```python
# Visualisasi 3: Jenis Produk yang Paling Banyak Dibeli
plt.figure(figsize=(12, 8))
sns.countplot(x='product_type', data=data, order=data['product_type'].value_counts().index)
plt.title('Jumlah Pembelian Berdasarkan Jenis Produk')
plt.xlabel('Jenis Produk')
plt.ylabel('Jumlah Pembelian')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisasi 4: Distribusi Jumlah Pembelian
plt.figure(figsize=(10, 6))
sns.boxplot(x='quantity_x', data=data)
plt.title('Distribusi Jumlah Pembelian')
plt.xlabel('Jumlah Pembelian')
plt.show()
```

```python
# Visualisasi 5: Distribusi Harga Produk
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=20, kde=True)
plt.title('Distribusi Harga Produk')
plt.xlabel('Harga Produk')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Menghitung frekuensi pembelian untuk setiap harga produk
price_counts = data['price'].value_counts()

# Memilih 10 harga teratas
top_prices = price_counts.head(10)

# Visualisasi: Harga Produk yang Paling Sering Dibeli menggunakan Seaborn dan Matplotlib
plt.figure(figsize=(12, 8))
sns.barplot(x=top_prices.index, y=top_prices.values)
plt.xticks(rotation=45)
plt.title('Top 10 Most Purchased Product Prices')
plt.xlabel('Product Price')
plt.ylabel('Purchase Frequency')
plt.show()
```
